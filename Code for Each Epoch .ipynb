```python
### Actual Code
from datetime import datetime
import datetime
import time
start = time.time()
import psutil
from tabulate import tabulate
import pandas as pd
import numpy as np
import GPUtil
num_epoch = 100

stats = pd.DataFrame(columns=("gpu_id", "gpu_name", "gpu_load", "gpu_total_memory", "gpu_used_memory", "gpu_free_memory",
            "gpu_temperature","cpu_load","total_ram","available_ram","percent_ram","used_ram","free_ram","total_swap_memory","used_swap_memory",
            "free_swap_memory","percent_swap_memory","disk_total","disk_used","disk_free","disk_percent"))
def get_size(bytes, suffix="B"):
    factor = 1024
    for unit in ["", "K", "M", "G", "T", "P"]:
        if bytes < factor:
            return f"{bytes:.2f}{unit}{suffix}"
        bytes /= factor
for epoch in range(num_epoch):
    gpus = GPUtil.getGPUs()
    for gpu in gpus:
        stats.loc[epoch, ['gpu_id']] = gpu.id 
        stats.loc[epoch, ['gpu_name']] = gpu.name
        stats.loc[epoch, ['gpu_load']]= f"{gpu.load*100}%"
        stats.loc[epoch, ['gpu_total_memory']]= f"{gpu.memoryTotal}MB"
        stats.loc[epoch, ['gpu_used_memory']]= f"{gpu.memoryUsed}MB"
        stats.loc[epoch, ['gpu_free_memory']]= f"{gpu.memoryFree}MB"
        stats.loc[epoch, ['gpu_temperature']]= f"{gpu.temperature} °C" 
        stats.loc[epoch, ['cpu_load']]= f"{np.array(psutil.cpu_percent(percpu=True, interval=1)).sum()/psutil.cpu_count(logical=True)}%" 
        stats.loc[epoch, ['total_ram']]= get_size(float(psutil.virtual_memory()[0])) 
        stats.loc[epoch, ['available_ram']]= get_size(float(psutil.virtual_memory()[1])) 
        stats.loc[epoch, ['percent_ram']]= f"{psutil.virtual_memory()[2]}%" 
        stats.loc[epoch, ['used_ram']]= get_size(float(psutil.virtual_memory()[3])) 
        stats.loc[epoch, ['free_ram']]= get_size(float(psutil.virtual_memory()[4])) 
        stats.loc[epoch, ['total_swap_memory']]= get_size(float(psutil.swap_memory()[0])) 
        stats.loc[epoch, ['used_swap_memory']]= get_size(float(psutil.swap_memory()[1]))
        stats.loc[epoch, ['free_swap_memory']]= get_size(float(psutil.swap_memory()[2])) 
        stats.loc[epoch, ['percent_swap_memory']]= f"{psutil.swap_memory()[3]}%" 
        stats.loc[epoch, ['disk_total']]= get_size(float(psutil.disk_usage('/')[0])) 
        stats.loc[epoch, ['disk_used']]= get_size(float(psutil.disk_usage('/')[1])) 
        stats.loc[epoch, ['disk_free']]= get_size(float(psutil.disk_usage('/')[2])) 
        stats.loc[epoch, ['disk_percent']]= f"{psutil.disk_usage('/')[3]}%" 

end= time.time()
stopWatch = end-start
print('Total Processing  Time (Second):',stopWatch)
```

* EN:

If you run this code in every Epoch process, you can save your entire system's performance while doing business. I think you should because it's a good reason to wait 1 second longer per Epoch.
